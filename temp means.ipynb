```python
import gym
import energym

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from operator import add
```

```python
temp_variable = 4
power_variable = 85

n = 10
```

```python
###### Hot dry ######

env = gym.make('Eplus-discrete-hot-v1')


indoor_temp_mean = [0 for i in range(35040)]
total_power_mean = [0 for i in range(35040)]


for i in range(n):
    obs = env.reset()

    in_temp = []
    rewards = []
    total_power = []

    done = False
    current_month = 0

    while not done:
        a = env.action_space.sample()
        obs, reward, done, info = env.step(a)

        in_temp.append(info['temperature'] + temp_variable*np.random.randn())
        total_power.append(info['total_power'] + power_variable*np.random.randn())
        rewards.append(reward)

        if info['month'] != current_month: 
            current_month = info['month']
            print('Reward: ', sum(rewards), info)
            print()
    print('Episode ', i, 'Mean reward: ', np.mean(rewards), 'Cumulative reward: ', sum(rewards))
    print()

    indoor_temp_mean = list(map(add, indoor_temp_mean, in_temp))
    total_power_mean = list(map(add, total_power_mean, total_power))

indoor_temp_mean = [i/n for i in indoor_temp_mean]
total_power_mean = [i/n for i in total_power_mean]

env.close()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12,4))

axes[0].plot(in_temp)
axes[0].set_title('Original indoor temperature')
axes[0].set_xlabel('timestep')
axes[0].set_ylabel('temperature ($^\circ$C)')

axes[1].plot(total_power)
axes[1].set_title('Original power consumption')
axes[1].set_xlabel('timestep')
axes[1].set_ylabel('power units')

plt.tight_layout()

plt.savefig('project_images/original_hot_indoortemp_and_power.png')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12,4))

axes[0].plot(indoor_temp_mean)
axes[0].set_title('Mean indoor temperature')
axes[0].set_xlabel('timestep')
axes[0].set_ylabel('temperature ($^\circ$C)')

axes[1].plot(total_power_mean)
axes[1].set_title('Mean power consumption')
axes[1].set_xlabel('timestep')
axes[1].set_ylabel('power units')

plt.tight_layout()

plt.savefig('project_images/mean_hot_indoortemp_and_power.png')
```

```python
hot_means = pd.DataFrame(list(zip(indoor_temp_mean, total_power_mean)), columns = ['indoor_temp_mean', 'total_power_mean'])

hot_means.to_csv('mean_runs/hot_means.csv')
```
